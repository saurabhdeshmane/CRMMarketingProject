# Function to Recommend Top 5 Similar Songs

```python
def getMostSimilarUser(msno_user1):
   return user_user_sim_matrix.loc[msno_user1].sort_values(ascending=False).index[2]
```

```python
print(getMostSimilarUser(2))
```

```python
def recommend(msno_user1,msno_user2,indexstart,indexend):
  songs_listened_user1 = set(user_song_matrix.loc[msno_user1].iloc[user_song_matrix.loc[msno_user1].to_numpy().nonzero()].index)
  songs_listened_user2 = set(user_song_matrix.loc[msno_user2].iloc[user_song_matrix.loc[msno_user2].to_numpy().nonzero()].index)
  songs_to_recommend = songs_listened_user1 - songs_listened_user2
  result= df.loc[df['song_id'].isin(songs_to_recommend), ['song_id', 'artist_name','genre_ids','msno']].drop_duplicates().set_index('song_id')[:5]
  index = range(indexstart,indexend)
  result['song_id'] = result.index
  result['index']=index
  result=result.set_index('index')
  return result
```

```python
d = recommend(1,799,0,5)
```

```python
final = pd.DataFrame()
```

```python
indexstart = 0
indexend = 5
```

```python
for i in listOfMsnos:
  user2 = getMostSimilarUser(i)
  res = recommend(i,user2,indexstart,indexend)
  final = pd.concat([final, res],axis=0)
  indexstart=indexstart + 5
  indexend = indexend + 5
```
